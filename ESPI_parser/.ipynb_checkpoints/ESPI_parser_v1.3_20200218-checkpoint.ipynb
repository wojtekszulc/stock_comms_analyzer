```python
def parse_date(soup):
    creation_date = soup.find('td',text=re.compile(r'Data sporz')).next_sibling.next_sibling.text.strip()
    creation_date = creation_date.replace("-","")
    return creation_date
```

```python
def parse_name(soup):
    name = soup.find('td',text="Nazwa emitenta").next_sibling.next_sibling.text
    name = name.replace("SPÓŁKA AKCYJNA", "SA").strip()
    name = name.replace("S.A.","SA")
    return name
```

```python
def parse_url(soup):
    contents = soup.find("div", {'class':'dane'})
    section_index = contents.find_all("table", {'class':'nDokument'})[0] # spis treści
    section_footer = contents.find_all("table", {'class':'nDokument'})[1] # stopka - informacje o firmie
    link_raport = section_index.find("a")['href']
    link_raport = link_raport[link_raport.find('#')+1:]
    start = contents.find('a',{'name':link_raport})
    content_table = start.parent.next_sibling.next_sibling
    contents_nTekst = content_table.find_all("tr", {'class':'nTekst'})
    total = [(el.text) for el in contents_nTekst]
    return ''.join(total)
```

```python
def parse_url_and_save(soup):
    contents = soup.find("div", {'class':'dane'})
    return contents
```

```python
path = './ESPI_htmls_zip/ESPI_htmls'
files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.html' in file:
            files.append(os.path.join(r, file))
files.sort()

dates=[]
names=[]
espi_info=[]
espi_bledy=[]
korupc=0
bledy=0
#espi = []
#espi.append(files[0])
#for f in files:
files = files[0:10000]
for f in files:
    url = str(f)
    #print(url)
    try:
        soup = BeautifulSoup(open(f), "html.parser")
        soup_text = parse_url(soup).lower()
        dates.append(parse_date(soup))
        names.append(parse_name(soup))
        espi_info.append(soup_text)
#        if soup_text.find("getback")>0:
#            print(soup_text[soup_text.find("getback")-50:soup_text.find("getback")+200])
#            korupc+=1
    except:
        #print("Błąd przy numerze: "+str(f))
        espi_bledy.append(f)
        bledy+=1
#     if len(espi_info)==1000:
#         break

print(len(files))
print(bledy)
print(bledy/len(files))
#print(espi_bledy)
print(korupc)
```

```python
data = pd.concat(
            [temp,
             -temp.groupby('Name')['High'].diff(periods=-1).rename('High+1'),
             -temp.groupby('Name')['High'].diff(periods=-2).rename('High+2'),
             -temp.groupby('Name')['High'].diff(periods=-3).rename('High+3'),
             -temp.groupby('Name')['High'].diff(periods=-4).rename('High+4'),
             -temp.groupby('Name')['High'].diff(periods=-5).rename('High+5')],
           axis=1)
data.columns=['High','High+1','High+2','High+3','High+4','High+5']
data['High+1_pct']=data['High+1']/data['High']
data['High+2_pct']=data['High+2']/data['High']
data['High+3_pct']=data['High+3']/data['High']
data['High+4_pct']=data['High+4']/data['High']
data['High+5_pct']=data['High+5']/data['High']
```

```python
def stemming_func(tokenized_t):
    stemmed_tekst = stemmer.stem(tokenized_t, parser)
    return [list(stemmed_tekst[i][1].keys())[0] for i in range(len(stemmed_tekst)) if len(list(stemmed_tekst[i][1].keys()))>0]
```
